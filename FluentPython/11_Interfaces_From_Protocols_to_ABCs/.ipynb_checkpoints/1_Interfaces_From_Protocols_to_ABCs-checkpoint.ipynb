> **鸭子类型** 实现的是特征动态协议     
> **协议** 即非正式接口，只由文档和约定定义，无法施加限制     
> **抽象基类** 使接口更明确，能验证实现是否符合规定，能强制接口一致性     

### 1. Python 中的序列

> Python 会尽量支持基本协议，会特殊对待类序列的对象。

序列继承（`Sequence` 抽象基类和 `collections.abc` 中其他相关抽象基类的继承关系）     


```python
class Foo:
    def __getitem__(self, pos):
        return range(0, 30, 10)[pos]

f = Foo()
```

```python
f[1]
```

```python
for i in f:
    print(i)
```

```python
20 in f, 15 in f
```

**如果类中没有定义 `__iter__` 和 `__contains__` 方法，Python 会自动调用 `__getitem__` 方法，使迭代和 `in` 操作符可用**     
**所以之前在鸭子类型中，只要定义了 `__getitem__` 和 `__len__` 方法就是序列，这只是一种协议，而不是真正的接口，真正的接口是抽象基类**
