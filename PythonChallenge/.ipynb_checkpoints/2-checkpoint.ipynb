````python
import requests
````

````python
url = 'http://www.pythonchallenge.com/pc/def/ocr.html'
````

````python
r = requests.get(url)
````

````python
import re
````

````python
mess = ''.join(re.findall(r'-->(.*)-->', r.text, re.S))
````

````python
mess
````

````python
letters = ''.join(re.findall(r'[a-zA-Z0-9]', mess, re.S))
````

````python
letters
````
