````python
import zipfile, re
````

````python
zfile = zipfile.ZipFile('channel.zip')
````

````python
start_index = '90052'
````

````python
next_index = start_index
while True:
    text = zfile.read(next_index + '.txt').decode()
    print(text)
    try:
        next_index = re.search(r'\d+', text).group()
    except:
        break
        
````

**看到找内容是没有用的，提示要找压缩文件注释**

````python
next_index = start_index
comments = ''
while True:
    text = zfile.read(next_index + '.txt').decode()
    print(text)
    comment = zfile.getinfo(next_index + '.txt').comment.decode()
    comments += comment
    print(comment)
    try:
        next_index = re.search(r'\d+', text).group()
    except:
        break
````
