*信息藏在图片的中间的马赛克区域*

````python
from PIL import Image
import requests
````

````python
img_url = 'http://www.pythonchallenge.com/pc/def/oxygen.png'
````

````python
with open('oxygen.png', 'wb') as f:
    f.write(requests.get(img_url).content)
````

````python
img = Image.open('oxygen.png')
````

````python
w, h = img.size
w, h
````

````python
img.getpixel((w-1, h-1))   # 4层图像，所以是4个像素值
````

````python
for i in range(w):
    print('=' * 10, i, '=' * 10)
    for j in range(h):
        print(j, img.getpixel((i, j)))
    
````

*从上面的结果可以看到，马赛克纵向的范围是 [43, 51]，因为这个范围里像素值一致，但是横向范围不好确定，只能看出从马赛克横向0开始，哪里结束慢慢找不合适，还是让代码解决*

````python
for i in range(w):
    if img.getpixel((i, 43)) != img.getpixel((i, 51)):
        print(i)
        break
````

*现在找到了马赛克的横向范围是 [0, 607]，然后在马赛克的范围里面做文章，可以马赛克区域发现四层图像像素值，前三个都是一致的，最后一个都是255，所以针对前三个的这个像素值，尝试认为这个像素值是ASCII码，用chr方法得到对应的ASCII字符*

````python
info = ''.join(chr(img.getpixel((i, 47))[0]) for i in range(608))     # 纵坐标只要是[43, 51]之间的任意值就好
````

````python
info
````

````python
info1 = 's'
for v in info:
    if v != info1[-1]:
        info1 += v
````
